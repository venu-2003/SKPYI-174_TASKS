```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure
fig, ax = plt.subplots(figsize=(9, 6))

# Draw the three horizontal bands of the flag
ax.fill_between([0, 1], 2/3, 1, color="#FF9933")  # Saffron
ax.fill_between([0, 1], 1/3, 2/3, color="white")  # White
ax.fill_between([0, 1], 0, 1/3, color="#138808")  # Green

# Draw the Ashoka Chakra
r = 0.06  # radius of Ashoka Chakra
circle = plt.Circle((0.5, 0.5), r, color="navy", fill=False, lw=2)
ax.add_artist(circle)

# Draw the 24 spokes of the Ashoka Chakra
for i in range(24):
    angle = 2 * np.pi * i / 24
    x = 0.5 + r * np.cos(angle)
    y = 0.5 + r * np.sin(angle)
    ax.plot([0.5, x], [0.5, y], color="navy", lw=1)

# Set the limits and aspect ratio
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect(1)

# Remove axes and borders
ax.axis("off")

# Display the flag
plt.show()
```
